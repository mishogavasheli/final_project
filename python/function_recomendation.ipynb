## build the function

```python
def step_1(data): 
    name = input('type name of the dish OR pass: ')
    
    # check if we want pass the stap
    if name.strip(' ') == 'pass':
        step_2(data, data)
    else:
        # build a cycle wich will work before we will not put correct input 
        checkpoint = 0
        while (~checkpoint):
            # filter by recipe name
            result = data[data['name'].str.contains(name)]
            
            # check if we want pass the stap
            if name == 'pass':
                step_2(result, data)
                break
            # if no match after filtering try one more time
            elif (len(result) == 0): 
                print('\n no match try again \n')
                name = input('type name of the dish ')
            else:    
                step_2(result, data)
                break
    return 
```

```python
def step_2(data, step_1_data):
    ingredient = input('type ingredient OR pass: ')
    
    
    if ingredient.strip(' ') == 'pass':
        step_3(data, data, step_1_data)
    else: 
        
        
        ingredient = ingredient.split(', ')
        result = data
        
        #
        if len(ingredient) > 1:
            for i in range(len(ingredient)): 
                 result = result[result['ingredients'].str.contains(ingredient[i])]
            
             
            if len(result) == 0: 
                step_1(step_1_data)
            else: 
                step_3(result, data, step_1_data)
        else: 
            
            result = result[result['ingredients'].str.contains(ingredient[0])]
            if len(result) == 0 : 
                step_1(step_1_data)
            else: 
                step_3(result, data, step_1_data)
```

```python
def step_3(data, step_2_data, step_1_data): 
    ingredient = input('type ingredient which you dont like OR pass: ')
    result = data
    
    
    if ingredient == 'pass':
        step_4(data, data, step_2_data, step_1_data)
    else:
        
        ingredient = ingredient.split(', ')        
        if len(ingredient) > 1: 
            for i in range(len(ingredient)): 
                result = result[~result['ingredients'].str.contains(ingredient[i])]
            if len(result) > 0: 
                step_4(result, data, step_2_data, step_1_data)
            else: 
                step_2(step_2_data, step_1_data)
        else: 
            result = result[~result['ingredients'].str.contains(ingredient[0])]
            if len(result) < 0 : 
                step_2(step_2_data, step_1_data)
            else: 
                step_4(result, data, step_2_data, step_1_data)
```

```python
def step_4(data, step_3_data, step_2_data, step_1_data): 
    vegetarian = input('vegetarian? yes OR pass: ')
    vegetarian = vegetarian.strip(' ')
    result = data
    if vegetarian == 'pass':
        step_5(data, data, step_3_data, step_2_data, step_1_data)
    else: 
        if vegetarian == 'yes': 
            result = result[result['vegetarian'] == True]
        else: 
            result = result[result['vegetarian'] == False]
        if len(result) > 0: 
            step_5(result, data, step_3_data, step_2_data, step_1_data)
        else: 
            step_3(step_3_data, step_2_data, step_1_data)
```

```python
def step_5(data, step_4_data, step_3_data, step_2_data, step_1_data): 
    difficulty = input('choose: easy, medium, difficult OR pass: ')
    difficulty = difficulty.strip(' ')
    if difficulty == 'pass':
        step_6(data, data, step_4_data, step_3_data, step_2_data, step_1_data)
    else: 
        result = data[data['difficulty'] == difficulty]
        if len(result) > 0: 
            step_6(result, data, step_4_data, step_3_data, step_2_data, step_1_data)
        else: 
            step_4(step_4_data, step_3_data, step_2_data, step_1_data)
```

```python
def step_6(data, step_5_data, step_4_data, step_3_data, step_2_data, step_1_data): 
    speed = input('choose: quick OR pass: ')
    speed = speed.strip(' ')
    if speed == 'pass':
        step_7(data, data, step_5_data, step_4_data, step_3_data, step_2_data, step_1_data)
    else: 
        result = data[data['quick'] == speed]
        if len(result) > 0: 
            step_7(result, data, step_5_data, step_4_data, step_3_data, step_2_data, step_1_data)
        else: 
            step_5(step_5_data, step_4_data, step_3_data, step_2_data, step_1_data)
```
